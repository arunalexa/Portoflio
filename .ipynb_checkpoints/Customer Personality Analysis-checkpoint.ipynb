#### Handling Missing Values
The Income column contains 24 missing values out of 2240 rows (~1%).
Since this proportion is small and Income is a key feature for segmentation,
these rows were removed to avoid bias from imputation.

```python
df.columns
```

```python
df["Age"] = 2026 - df["Year_Birth"]
```

```python
df["Age"].describe()
```

```python
df = df[df["Age"] <= 100]
df["Total_Spending"] = (
    df["MntWines"] +
    df["MntFruits"] +
    df["MntMeatProducts"] +
    df["MntFishProducts"] +
    df["MntSweetProducts"] +
    df["MntGoldProds"]
)
df["Total_Purchases"] = (
    df["NumWebPurchases"] +
    df["NumCatalogPurchases"] +
    df["NumStorePurchases"]
)
df["Total_Accepted"] = (
    df["AcceptedCmp1"] +
    df["AcceptedCmp2"] +
    df["AcceptedCmp3"] +
    df["AcceptedCmp4"] +
    df["AcceptedCmp5"]
)
```

```python
## Preparing for K-means clustierng

features = df[["Income", "Total_Spending", "Total_Purchases", "Recency"]]

features.isnull().sum()
```

```python
# Since income has 24 missing values, we are removing them first
df_model = df.dropna(subset=["Income"]).copy()

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

features = df_model[["Income", "Total_Spending", "Total_Purchases", "Recency"]]

scaler = StandardScaler()
scaled = scaler.fit_transform(features)

kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
df_model["Cluster"] = kmeans.fit_predict(scaled)
 
```

```python
#sanity checks
df_model.isnull().sum()[df_model.isnull().sum() > 0]
df_model["Cluster"].value_counts()
```

```python
df_model.groupby("Cluster")[["Income","Total_Spending","Total_Purchases","Recency","Age"]].mean()
```

```python
import matplotlib.pyplot as plt

df_model.boxplot(column="Total_Spending", by="Cluster")
plt.title("Total Spending by Cluster")
plt.suptitle("")
plt.xlabel("Cluster")
plt.ylabel("Total Spending")
plt.show()
```

```python
#“The elbow curve shows a sharp decrease in inertia from k=1 to k=3, after which the rate of decrease slows significantly.
#This suggests that 3 clusters capture most of the meaningful structure in the data.

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(scaled)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia, marker="o")
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()
```

```python
from sklearn.metrics import silhouette_score

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(scaled)
    score = silhouette_score(scaled, labels)
    print(f"k={k}, silhouette={score:.3f}")
```

```python
# Response rate per Segment
df_model.groupby("Cluster")["Response"].mean()
```

### Average response rate for each cluster. 
The results show significant variation across segments:
Cluster 0 → 4.7%
Cluster 1 → 29.3%
Cluster 2 → 14.4%
Cluster 3 → 15.3%

Cluster 1 exhibits a substantially higher response rate compared to the other segments. In contrast, Cluster 0 shows very low responsiveness.

The behavioral segmentation appears meaningful, as clusters differ not only in spending patterns but also in campaign sensitivity. This suggests that certain behavioral profiles are more likely to engage with marketing campaigns.

Business Implication:

Future campaigns could prioritize Cluster 1 customers to maximize conversion efficiency. Cluster 0 may require alternative engagement strategies or lower cost communication channels.

Campaign Response by Segment

The overall campaign response rate across the dataset is 15.05%. However, response rates vary significantly across customer segments:

Cluster 1 exhibits a response rate of 29.3%, nearly double the overall average. This indicates that customers in this segment are highly receptive to marketing campaigns.

In contrast, Cluster 0 shows a response rate of only 4.7%, suggesting low campaign sensitivity.

Clusters 2 and 3 have response rates close to the overall average, indicating moderate responsiveness.

Business Implication

The segmentation successfully identifies a high-value, high-responsiveness segment (Cluster 1). Prioritizing this segment for future campaigns could substantially improve marketing efficiency and ROI.

Lower responsiveness segments may require alternative engagement strategies or cost-optimized targeting approaches.

```python
# Logistic Regression to predict the effects on the response variable

from sklearn.model_selection import train_test_split

features = df_model[["Income", "Recency", "Total_Spending", "Total_Purchases", "Age"]]
target = df_model["Response"]

X_train, X_test, y_train, y_test = train_test_split(
    features,
    target,
    test_size=0.3,
    random_state=42,
    stratify=target   # important because response is imbalanced
)
```

```python
#scaling features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#Train logistic Regression 
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)

model.fit(X_train_scaled, y_train)
```

```python
#make predictions
y_pred = model.predict(X_test_scaled)
y_prob = model.predict_proba(X_test_scaled)[:,1]
```

```python
#evaluate model 
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix

print(classification_report(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, y_prob))
```

```python
import pandas as pd

coefficients = pd.DataFrame({
    "Feature": features.columns,
    "Coefficient": model.coef_[0]
})

coefficients.sort_values(by="Coefficient", ascending=False)
```

```python
threshold = 0.3
y_custom = (y_prob > threshold).astype(int)
print(classification_report(y_test, y_custom))
```
